## Dior

```python
df_dior.info()
```

```python
df_dior.set_index('date', inplace=True)
```

```python
# Ensure 'date' is in datetime format for df_dior
df_dior['date'] = pd.to_datetime(df_dior['date'])

# Now group by year, aggregating numerical values by mean
df_dior_yearly = df_dior.groupby(df_dior['date'].dt.to_period('Y')).mean().reset_index()

# Optional: Convert 'date' back to datetime format if needed
df_dior_yearly['date'] = df_dior_yearly['date'].dt.to_timestamp()
```

```python
# Plot each column with a different color
plt.figure(figsize=(14, 8)) 
plt.plot(df_dior['adj_close'], label='adj close', color='blue');
plt.plot(df_dior['close'], label='close', color='green');
plt.plot(df_dior['high'], label='high', color='red');
plt.plot(df_dior['low'], label='low', color='cyan');
plt.plot(df_dior['open'], label='open', color='magenta');
# plt.plot(df_dior['volume'], label='volume', color='yellow');

# Adding title and labels
plt.title('Dior Stock Prices Over Time');
plt.xlabel('date');
plt.ylabel('price');
plt.legend();

plt.xticks(rotation=45);
```
